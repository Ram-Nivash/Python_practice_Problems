```python
from google.colab import files
import re # re = regular expressions
```

```python
fil
key = list(fil.keys())
text = str(fil[key[0]])
text
```

```python
#example of re module
sentence = "Hi, my name is Mike and I like to eat Mike's chocalates."

regex_name = re.compile("Mike") # here we are initiating an re object containing Mike into it.  note = it's case sensative.
regex_name.findall(sentence)# findall method finds all the instances of that object in the given string
```

```python
regex_name.sub('MIKE',sentence)# sub method substitue the object with new object (in this case MIKE) in the given string(here it is sentence)
```

```python
sentence = "Hi, my name is Mike and I like to eat Muke's chocalates."
regex_name = re.compile(r'M\wke')# here we are using an special wild card = \w by adding this we are basically implying that this object is anythig which start with M(and anything in between) and then end with ke.
regex_name.findall(sentence)
```

```python
# pattern to replace
"\\n\\n1\\n00:05.080 --> 00:09.310\\"

pattern2replace = r"\\n\\n\d{1,3}\\n\d\d:\d\d.\d\d\d --> \d\d:\d\d.\d\d\d\\"
# now here we are using \d to represent any digit that will occure in that place and {1,3} this will tell python that we are defenetily searching for a digit but it not necessecry to have one digit, it can be one digit or two digit or three digit.
regex_timestamps = re.compile(pattern2replace)
regex_timestamps.findall(text)
```

```python
newtext = regex_timestamps.sub(' ',text)
newtext = newtext[10:-1]
newtext
```

```python
words = newtext.split()
print(len(words))
print(type(words))
```

```python
for wordi in range(len(words)):
  if len(words[wordi]) == 4 and not words[wordi][-1]==".":
    words[wordi] = words[wordi][0]+"***"

words
```

```python
cleantext = ' '.join(words) # join method is used on a string to join the objects of list by given expression in that string (in this case it is a space = ' ' )
cleantext
```

```python
# now download this string as txt file in our computer
with open('clean_transcript.txt','w') as f:
  f.write(cleantext)
files.download('clean_transcript.txt')
```

```python
## bonus : readiblity of scrambled words
bonus = "I have often wondered whether it's possible to read an entire book faster if the useless letters were removed."
bonus_words = bonus.split()
```

```python
import numpy as np
```

```python
# completely scramble all the letters
for wi in range(len(bonus_words)):
  thisword = bonus_words[wi]
  idx = np.random.permutation(len(thisword))
  bonus_words[wi]= ''.join([thisword[i] for i in idx])
' '.join(bonus_words)
```

```python
# scramble all the letters except for the first and the last word

bonus_words = bonus.split()

for wi in range(len(bonus_words)):
  #ignore short words
  if len(bonus_words[wi])<4:
    continue

  thisword = bonus_words[wi][1:-1]
  idx = np.random.permutation(len(thisword))
  bonus_words[wi]= bonus_words[wi][0]+''.join([thisword[i] for i in idx])+bonus_words[wi][-1]
' '.join(bonus_words)
```

```python
# replace with random letters
import string
allletters = string.ascii_lowercase

bonus_words = bonus.split()

for wi in range(len(bonus_words)):
  #ignore short words
  if len(bonus_words[wi])<4:
    continue

  thisword = bonus_words[wi][1:-1]
  idx = np.random.randint(0,len(allletters),len(thisword))
  bonus_words[wi]= bonus_words[wi][0]+''.join([allletters[i] for i in idx])+bonus_words[wi][-1]
' '.join(bonus_words)
```
